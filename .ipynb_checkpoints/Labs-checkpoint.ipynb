```python
import math

def task_10():
    try:
        num = int(input())
        for i in range (num//2):
            n1, n2 = math.sqrt(i), math.sqrt(num - i)
            if n1.is_integer() and n2.is_integer():
                print(str(int(n1)) + "^2+" + str(int(n2)) + "^2=" + str(num))
                return
        print("The number " + str(num) + " cannot be expressed as squares of integers")
    except:
        print("Wrong format")
```
